## Las variables
1. X = Cantidad de lápices
2. PAG = Pago que se realizará por los lápices

## Formulas
* $V = PAG = X * 0.85$
* $F = PAG = X * 0.90$

## Diagrama de flujo LucidChart

```python
# Para cuando sea verdadero (V)
# Determinando el pago por X cantidad de lápices
# Datos de entrada: X (Cantidad de lápices)
# Si X >= 1000, costo es 0.85; de lo contrario, 0.90
X = int(input("Ingrese la cantidad de lápices a comprar: "))

if X >= 1000:
    PAG = X * 0.85
else:
    PAG = X * 0.90

print("El pago total es: ", PAG)
print("Fin del programa")
print("Programa realizado por Samuel Martinez NL = 35")
```

```python
# Para cuando sea falso (F)
# Determinando el pago por X cantidad de lápices
# Datos de entrada: X (Cantidad de lápices)
# Si X >= 1000, costo es 0.85; de lo contrario, 0.90
X = int(input("Ingrese la cantidad de lápices a comprar: "))

if X >= 1000:
    PAG = X * 0.85
else:
    PAG = X * 0.90

print("El pago total es: ", PAG)
print("Fin del programa")
print("Programa realizado por Samuel Martinez NL = 35")
```
